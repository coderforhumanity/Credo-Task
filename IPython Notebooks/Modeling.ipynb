```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
# Reading data and storing in a dataframe
df = pd.read_csv("../Data/preprocessed_data.csv")
df
```

```python
# splitting the data into x and y variables
x = df.drop(columns=['impact'])
y = df['impact']
```

```python
# splitting the data into train test datasets
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# Implementing Logistic Regression
model = LogisticRegression()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
```

```python
# Implementing Decision Tree Classifier
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
```

```python
# Implementing Random Forest Classifier
model = RandomForestClassifier()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
```

```python
# Implementing KNN Classifier
model = KNeighborsClassifier(n_neighbors=2)
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
```

```python
# Implementing Support Vector Classifier
model = SVC(kernel='linear')
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
print(confusion_matrix(ytest, ypred))
print(classification_report(ytest, ypred))
```

As we could see from the above results that by applying KNN classifier, we can get the maximum recall value for 1 -High impact (as this is our class of importance), therefore we should work towards improvement of the KNN model in order to classify the 1- High impact in a better way. Also we should improve the random forest model as it gives the highest precision and recall value in terms of 1-High impact class. Therefore the random forest model is the most stable model.
