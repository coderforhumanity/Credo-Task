This is the data cleaning notebook. We took the data into a dataframe and dropped a few columns which we though may not be important. Moreover we dropped the records where we had '?'. We could have imputed the '?' records with suitable values but for the POC purpose we thought not to do so as it is much time consuming process.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/data.csv')
df
```

```python
df['impact'].value_counts()
```

```python
df.info()
```

```python
df.drop(columns = ['Unnamed: 0', 'opened_time','notify', 'problem_ID', 'change_request', 'created_at', 'updated_at'], axis = 1, inplace=True)
df
```

```python
df = df.replace('?', pd.NaT)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df['impact'].value_counts()
```

```python
df.to_csv('../Data/cleaned_data.csv', index = False)
```
